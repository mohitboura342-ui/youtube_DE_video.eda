## Literature Review on EDA for YouTube Trending Data

Exploratory Data Analysis (EDA) is the first and most important step in understanding a dataset because it helps identify structure, missing values, outliers, relationships, and potential data quality issues before modeling begins. In YouTube trending video datasets, EDA is especially useful because variables such as views, likes, dislikes, comments, category, trending date, and publish time often show strong skewness, time-based changes, and category imbalance. [academic.oup](https://academic.oup.com/comjnl/article/66/1/35/6404045)

A recent study on YouTube trending videos showed that EDA can reveal meaningful patterns in engagement metrics and can improve downstream machine learning performance by clarifying which features are informative. The authors used statistical summaries and visual analysis to study a dataset of 40,950 videos collected over 205 days and reported that views and likes were strongly positively correlated, while Entertainment was the most dominant category among trending videos. Another ACM study on YouTube trending data also examined the factors associated with trending-page inclusion and reinforced the idea that popularity signals such as views and engagement play a central role in trending behavior. [dl.acm](https://dl.acm.org/doi/10.1145/3572647.3572668)

## Data Inspection and Cleaning

The literature consistently recommends starting with schema inspection using tools such as `read_csv()`, `info()`, and `describe()` to verify column types, detect missing values, and understand numeric distributions. Pandas documentation supports `read_csv()` as the standard method for loading CSV files, while `describe()` and `info()` are widely used for quick inspection of numerical and structural properties. For social-media datasets, this step is critical because text fields may contain empty strings, dates may require conversion, and large files may need encoding checks before full analysis. [pandas.pydata](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html)

## Distribution and Outlier Analysis

EDA literature emphasizes that engagement variables such as views, likes, and comments are usually highly right-skewed rather than normally distributed. This means a small number of viral videos can dominate the range, making log transformation, histograms, and boxplots more informative than raw-value summaries alone. Outlier detection is therefore not just a cleaning step but also an analytical step, because unusually high values may represent true viral behavior rather than errors. [ebsco](https://www.ebsco.com/articles/biography/0edba6f9-7465-5129-80b8-4eb319d3d7e5/youtube-trending-videos-boosting-machine-learning-results-using-exploratory-data-analysis)

## Correlation and Feature Interaction

Correlation analysis is another core theme in EDA studies of YouTube data. Prior research reports strong positive relationships between views and likes, and often between views and comments as well. Heatmaps, scatterplots, and pairwise visualizations are commonly recommended to explore these associations and to check whether multicollinearity may affect later modeling. However, the literature also warns that correlation does not capture non-linear effects, so scatterplots and transformed axes are necessary for a fuller interpretation. [medium](https://medium.com/@a7m7da7e3/ahands-on-guide-to-exploratory-data-analysis-eda-with-pandas-matplotlib-and-seaborn-08da242e38c9)

## Time-Series and Category Imbalance

Because trending datasets include a date dimension, many studies recommend aggregating by day or week to detect temporal spikes, seasonality, and platform-driven patterns. This is especially relevant when the same video appears across multiple trending days. Category imbalance is another repeated finding: entertainment and music-related content often appear more frequently than smaller categories, which can bias descriptive analysis and predictive models if not handled carefully. [dl.acm](https://dl.acm.org/doi/fullHtml/10.1145/3572647.3572668)

## Practical EDA Tools

The literature strongly supports Python-based workflows using Pandas and Seaborn, especially in Jupyter notebooks, because they make EDA reproducible and visually clear. Pandas is used for loading, cleaning, grouping, and summarizing data, while Seaborn is commonly used for histograms, boxplots, scatterplots, and heatmaps. For academic reporting, this combination is repeatedly recommended because it produces both statistical summaries and publication-ready visualizations. [rockborne](https://rockborne.com/insights/analysis-python-pandas/)

## Conclusion

Overall, the literature shows that EDA for YouTube trending datasets should focus on schema validation, missing-value checks, distribution analysis, correlation study, outlier handling, time-based aggregation, and class imbalance assessment. Your proposed workflow matches established best practice in the literature and is well suited for producing both descriptive insights and a foundation for later predictive modeling. [academic.oup](https://academic.oup.com/comjnl/article/66/1/35/6404045)

```python
import matplotlib.pyplot as plt

# Correct weekday order
weekday_order = ['Monday', 'Tuesday', 'Wednesday',
                 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Count videos published on each weekday
weekday_counts = (
    df['publish_weekday']
    .value_counts()
    .reindex(weekday_order)
    .fillna(0)
)

plt.figure(figsize=(12,6))

bars = plt.bar(
    weekday_counts.index,
    weekday_counts.values,
    edgecolor='black'
)

# Average line
avg = weekday_counts.mean()
plt.axhline(avg, color='red', linestyle='--',
            linewidth=2, label=f'Average = {avg:.1f}')

# Add Count + Percentage
total = weekday_counts.sum()

for bar in bars:
    height = bar.get_height()
    percentage = (height / total) * 100

    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 5,
        f'{int(height)}\n({percentage:.1f}%)',
        ha='center',
        fontsize=9
    )

# Statistics Box
stats = (
    f"Total Videos : {int(total)}\n"
    f"Best Day : {weekday_counts.idxmax()}\n"
    f"Highest : {int(weekday_counts.max())}\n"
    f"Least Day : {weekday_counts.idxmin()}\n"
    f"Lowest : {int(weekday_counts.min())}\n"
    f"Average : {avg:.1f}"
)

plt.text(
    0.98, 0.98,
    stats,
    transform=plt.gca().transAxes,
    ha='right',
    va='top',
    fontsize=10,
    bbox=dict(facecolor='white', alpha=0.9)
)

plt.title("Trending Videos by Publish Weekday", fontsize=16)
plt.xlabel("Weekday")
plt.ylabel("Number of Trending Videos")
plt.grid(axis='y', alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()
```

**videos published on weekdays—especially Thursday and Friday—were more likely to become trending than those published on weekends.**
